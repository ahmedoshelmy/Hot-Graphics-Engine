```python
from skimage import io, img_as_float
import numpy as np
```

```python
BASE = "./Flashlight_2"
path_ao = "_ao.png"
path_metallic = "_Metallic.png"
```

```python
ao = img_as_float(io.imread(BASE+path_ao))# / 256.0
metallic = img_as_float(io.imread(BASE+path_metallic)[:,:,0]) 
n,m = ao.shape
r = np.ones((n,m))
```

```python
#r_ao_m
# roughness = 0
stacked = np.stack((r, ao, metallic), axis=2)
stacked = img_as_float(stacked)
print(stacked.shape)
io.imshow(stacked)
```

```python
image_uint8 = (stacked * 255).astype(np.uint8)

io.imsave(BASE+"r_ao_m.png", image_uint8)
```
